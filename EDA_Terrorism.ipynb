```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/globalterrorismdb_0718dist.csv",encoding='latin1')
print(df.head())
```

```python
# Group the data by country and calculate the total number of attacks.
attacks_by_country = df.groupby('country')['eventid'].count().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
```

```python
sns.barplot(x=attacks_by_country.head(10).values, y=attacks_by_country.head(10).index)
plt.xlabel('Number of Attacks')
plt.ylabel('Country')
plt.title('Top 10 Countries with the Most Terrorist Attacks')
plt.show()
```

```python
# Group the data by region and calculate the total number of attacks.
attacks_by_region = df.groupby('region')['eventid'].count().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
```

```python
sns.barplot(x=attacks_by_region.head(5).values, y=attacks_by_region.head(5).index)
plt.xlabel('Number of Attacks')
plt.ylabel('region')
plt.title('Top 5 Regions with the Most Terrorist Attacks')
plt.show()
```

```python
pivot_table = df.pivot_table(values='eventid', index='iyear', columns='region', aggfunc='count')
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='PuBuGn')
plt.xlabel('region')
plt.ylabel('iyear')
plt.title('Heatmap of Terrorist Attacks by Year and Region')
plt.show()
```
